## Load

```python
import pandas as pd
df = pd.read_pickle("geolife.pkl")
df
```

```python
df["prevlat"] = df.lat.shift(1)
df["prevlon"] = df.lon.shift(1)
df["prevalt"] = df.alt.shift(1)
df["prevtime"] = df.time.shift(1)
df = df.fillna(0)
df["v"] = abs((df.lat-df.prevlat)+(df.lon-df.prevlon)+(df.alt-df.prevalt))/(df.time-df.prevtime+1e-9)
df
```

```python
train_data = df[df.label!=0]
test_data = df[df.label==0]
display(train_data)
display(test_data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(train_data[["time","lat","lon","alt","user","v"]],train_data.label, shuffle=False)
X_test = test_data[["time", "lat","lon","alt","v"]]
display(X_train)
display(y_train)
```

## 랜덤포레스트

```python
rf_classifier = RandomForestClassifier(n_estimators = 18)
rf_classifier.fit(X_train,y_train)
```

```python
display(rf_classifier.score(X_train, y_train))
display(rf_classifier.score(X_valid,y_valid))
display(rf_classifier.predict(X_valid))
```

```python
rf_classifier.predict(X_test)
```

## LogisticRegression

```python
logreg_classifier = LogisticRegression()
logreg_classifier.fit(X_train,y_train)
```

```python
display(logreg_classifier.score(X_train,y_train))
display(logreg_classifier.score(X_valid,y_valid))
display(logreg_classifier.predict(X_test))
```
